### 1.1: Escribe una función que te devuelva una matriz de números aleatorios entre 1 y 100 con 2 decimales. El número de filas y columnas de la matriz será determinado por sendos parámetros. Utiliza la función para crear una matriz de 10 filas y 7 columnas.
<font color="#9E1B32">**(No utilices NUMPY ni PANDAS en este apartado 1.1, en los demás puedes usarlo)**</font>


```python
#Importo el módulo random para obtener los números aleatorios y dentro de random elijo random.uniform para poder poner número de decimales que quiero
import random

#Creo la función matriz aleatoria
def matriz_aleatoria(filas, columnas):
    matriz=[]
    for i in range(filas):
       fila=[]
       for j in range(columnas):
          fila.append(round(random.uniform(1, 100), 2))
       matriz.append(fila)

    return matriz

#Ejecuto la función y le asigno el nombre MA y lo muestro en pantalla de manera más visual
MA= matriz_aleatoria(10,7)
for fila in MA:
    print(fila)
```

### 1.2: Normaliza los elementos de la matriz anterior, de modo que todos los valores se encuentren entre 0 y 1


```python

#Importo librería numpy
import numpy as np
#Utilizo la biblioteca scikit-learn e importo MinMaxScaler para hacer la normalización de la matriz
from sklearn.preprocessing import MinMaxScaler

#Creo el objeto MinMaxScaler
scaler = MinMaxScaler()

#Ajusto y transformo los datos
MA_normalizada = scaler.fit_transform(MA)
#Muestro en pantalla la matriz normalizada
print(MA_normalizada)


```

### 1.3: A partir de la matriz normalizada, filtra todos aquellos valores comprendidos entre 0.10 y 0.90
El resultado debe ser un vector con dichos valores ordenados descendentemente

```python
#Defino Matriz aleatoria normalizada filtrada vacía para guardar el resultado

MA_normalizada_filtrada = []

#Hago bucles para recorrer la matriz

for i in range(len(MA_normalizada)):
    for j in range(len(MA_normalizada[i])):
        if MA_normalizada[i][j] > 0.90 or MA_normalizada[i][j] < 0.10:
            MA_normalizada[i][j] = 0
        else:
            MA_normalizada_filtrada.append(MA_normalizada[i][j])

#Ordeno de manera descendente la matriz normalizada filtrada con el comando sorted

MA_normalizada_filtrada_ordenada=sorted(MA_normalizada_filtrada, reverse=True)

#Muestro en pantalla la matriz normalizada filtrada y ordenada de manera descendente
print(MA_normalizada_filtrada_ordenada)
```

```python
df[["RetrasoSalida"]].plot(kind = 'box')
plt.title('Outliers en RetrasoSalida')
plt.show()
```

Para quitar los outliers voy a usar el rango intercuartílico **(IQR)**

```python
#Calculo primero los cuartiles para la variable "RetrasoSalida"
Q1= df['RetrasoSalida'].quantile(0.25)
Q3 = df['RetrasoSalida'].quantile(0.75)
IQR = Q3 - Q1

# Defino los límites para los outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtro los valores atípicos
df_sinoutliers = df[(df['RetrasoSalida'] >= limite_inferior) & (df['RetrasoSalida'] <= limite_superior)]
```

```python
df_sinoutliers["RetrasoSalida"].plot(kind = 'box')
plt.title('RetrasoSalida después de eliminar Outliers')
plt.show()
```

```python
df_sinoutliers["RetrasoLlegada"].plot(kind = 'box')
plt.title('Outliers en RetrasoLlegada')
plt.show()
```

```python
#Calculo primero los cuartiles para la variable "RetrasoLlegada"
Q1= df_sinoutliers['RetrasoLlegada'].quantile(0.25)
Q3 = df_sinoutliers['RetrasoLlegada'].quantile(0.75)
IQR = Q3 - Q1

# Defino los límites para los outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtro los valores atípicos
df_sinoutliers_2 = df_sinoutliers[(df_sinoutliers['RetrasoLlegada'] >= limite_inferior) & (df_sinoutliers['RetrasoLlegada'] <= limite_superior)]
```

```python
df_sinoutliers_2["RetrasoLlegada"].plot(kind = 'box')
plt.title('RetrasoLlegada después de eliminar Outliers')
plt.show()
```

Retrasos medios de salida y llegada, los calculo con la función promedio mean

```python
Retraso_MEDIO_salida = df_sinoutliers_2['RetrasoSalida'].mean()
print("Retraso MEDIO de salida:", Retraso_MEDIO_salida, "minutos")

Retraso_MEDIO_llegada = df_sinoutliers_2['RetrasoLlegada'].mean()
print("Retraso MEDIO de llegada:", Retraso_MEDIO_llegada, "minutos")
```

Comportamiento de las compañías aéreas en cuanto a retrasos de llegada y salida medios. Saco un gráfico para ver estos datos de una manera más clara


```python
# Calcular el retraso de salida medio por compañía
retraso_medio_salida = df_sinoutliers_2.groupby("Compañía")["RetrasoSalida"].mean()
#Calcular el retraso de llegada medio por compañía
retraso_medio_llegada = df_sinoutliers_2.groupby("Compañía")["RetrasoLlegada"].mean()
plt.figure(figsize=(10, 6))
```

```python
retraso_medio_salida.plot(kind='bar', color='green')

plt.title('Retraso Medio de Salida por Compañía')
plt.xlabel('Compañía')
plt.ylabel("Retraso Medio de Salida")
plt.show()
```

```python
retraso_medio_llegada.plot(kind='bar', color='green')

plt.title('Retraso Medio de Llegada por Compañía')
plt.xlabel('Compañía')
plt.ylabel("Retraso Medio de Llegada")
plt.show()
```

¿Qué aeropuerto tiene el mayor número de retrasos de salida? Para calcularlo agrupo por AeropuertoOrigen y hago la suma de los RetrasosSalida, luego los ordeno de mayor a menor.

```python
retrasos_por_aeropuerto = df_sinoutliers_2.groupby("AeropuertoOrigen")["RetrasoSalida"].sum()
retrasos = retrasos_por_aeropuerto.sort_values(ascending=False)
```
